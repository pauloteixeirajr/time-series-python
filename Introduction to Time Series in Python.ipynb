## Plotting the Data

```python
import matplotlib.pyplot as plt
```

```python
df_comp.spx.plot(title='S&P500 Prices', figsize=(20, 5))
plt.show()
```

```python
df_comp.ftse.plot(title='FTSE100 Prices', figsize=(20, 5))
plt.show()
```
